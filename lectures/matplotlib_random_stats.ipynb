```python
from scipy.stats import norm

x = np.random.normal(loc = 1.0, scale = 0.2, size = 100)

y = norm.pdf(x, loc = 1.0, scale = 0.2)

plt.hist(x, density = True, alpha = 0.5)
plt.scatter(x, y, color = "red")

```

```python
y_target = 2 * x + 5

plt.scatter(x, y_target, color = "green")
```

# Linear regression

```python
#np.random.seed(101)

#x[:, None]               # shape: (3, 1) <- 2d (single column matrix)
#x[:, np.newaxis]         # shape: (3, 1) <- a meaningful alias to None
#x.reshape(-1, 1)         # shape: (3, 1)

# number of samples
N = 100

# Generate points
X = np.random.normal(loc=2.5, scale=2.0, size=N)
X = X[:, np.newaxis]
print(X.shape)

# generate target function
#y = 2.2 * X + 1 + 0.8 * np.random.randn(N, 1)
#y = 2 * X**2 + -15.2 * X + 1 + 0.8 * np.random.randn(N, 1)
norm_noise = np.random.normal(loc=0, scale=0.1, size=N)
norm_noise = norm_noise.reshape(-1, 1)
y = np.sin(X) + norm_noise


# generate x_0 
X0 = np.ones((N, 1))
print(X0.shape)

# add bias x_0
Xnew = np.hstack((X0, X))
print(Xnew.shape)

# Xnew

# find solution vector w = (X^t X)^-1 X^T y
w = np.dot(np.linalg.inv(np.dot(Xnew.T, Xnew)), np.dot(Xnew.T, y))

print(w)

plt.plot(X, y, '.')

xmin = X.min()
xmax = X.max()

plt.plot([xmin, xmax], [xmin * w[1] + w[0], xmax * w[1] + w[0]])


```

```python
# compute in-sample error
E_in = (X * w[1] + w[0] - y)**2

plt.plot(X, E_in, '.')
```
