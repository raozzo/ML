# REGRESSIONE LINEARE

```python
import numpy as np
import matplotlib.pyplot as plt

N = 100

X_sample = np.random.normal(loc=0.5, scale= 0.7, size= N)
X0 = np.ones(N)
X = np.stack((X0, X_sample), axis = 1)
Y = 2*X_sample + np.random.normal(loc=0.6, scale=0.7, size=N)

plt.scatter(X[:,1], Y)

w = np.dot(np.linalg.inv(np.dot(X.transpose(), X)), np.dot(X.transpose(), Y))
print(w)

x_grid = np.linspace(start=-1, stop=2, num=100)
y_grid = x_grid*w[1]+w[0]
plt.plot(x_grid, y_grid, color = 'red')
```

# PERCETTRONE CON FUNZIONE D'ERRORE

```python
def error_w(X,Y,w):
    error = sum(np.square(Y-np.dot(X,w)))/2
    return error
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2510)
N = 100

X1_1 = np.random.normal(loc=1, scale=0.5, size=N)
X2_1 = 2*X1_1 + np.random.normal(loc=0.5, scale=0.3, size=N)
Y_1 = np.ones(N)

X1_2 = np.random.normal(loc=1.5, scale=0.5, size=N)
X2_2 = 2*X1_2 + np.random.normal(loc=0.5, scale=0.3, size=N) - 2.4
Y_2 = -np.ones(N)

#plt.scatter(X1_1,X2_1, color = 'green')
#plt.scatter(X1_2, X2_2, color = 'red')

w = np.random.uniform(low=0, high =3, size=3)

plt.plot([np.min(X1_1), np.max(X1_2)], [np.min(X1_1)*(-w[1])/w[2]+(-w[0]/w[2]),np.max(X1_2)*(-w[1])/w[2]+(-w[0]/w[2])], color = 'black')

X0 = np.ones(2*N)
X1 = np.concatenate((X1_1, X1_2))
X2 = np.concatenate((X2_1, X2_2))
X = np.stack((X0,X1,X2), axis=1)

Y = np.concatenate((Y_1, Y_2))

plt.scatter(X[:,1], X[:,2], c=Y)

epochs = 0
max_epochs = 100

errori = np.zeros(max_epochs)
numero_errori = np.zeros(max_epochs)

while(epochs<max_epochs):
    num_errors = 0
    errori[epochs] = error_w(X,Y,w)
    print("Epoca numero: ", epochs)
    print("Errore: ", errori[epochs])
    for i in range(Y.shape[0]):
        x = X[i,:]
        y = Y[i]
        dot_prod = y*np.dot(x,w)
        if(dot_prod<0):
            num_errors +=1
            w = w + x*y
    numero_errori[epochs] = num_errors
    print("Numero errori:  ", numero_errori[epochs])     
    if(num_errors==0):
        break
    epochs+=1


print("L'errore dell'ultima epoca e'", errori[-1])
print("Il numero di errori dell'ultima epoca e'", numero_errori[-1])
plt.plot([np.min(X1_1), np.max(X1_2)], [np.min(X1_1)*(-w[1])/w[2]+(-w[0]/w[2]),np.max(X1_2)*(-w[1])/w[2]+(-w[0]/w[2])], color = 'blue')       
plt.scatter(X[:,1], X[:,2], c=Y)       
```

```python
fig,axs = plt.subplots(1,2, figsize=(15,5), sharey=False)
axs[0].plot(range(sum(errori>0)), errori[errori>0])
axs[1].plot(range(sum(errori>0)), numero_errori[errori>0])
```

# GRADIENT DESCENT STOCASTICO

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2510)
N = 100

X1_1 = np.random.normal(loc=1, scale=0.5, size=N)
X2_1 = 2*X1_1 + np.random.normal(loc=0.5, scale=0.3, size=N)
Y_1 = np.ones(N)

X1_2 = np.random.normal(loc=1.5, scale=0.5, size=N)
X2_2 = 2*X1_2 + np.random.normal(loc=0.5, scale=0.3, size=N) - 2.4
Y_2 = -np.ones(N)

#plt.scatter(X1_1,X2_1, color = 'green')
#plt.scatter(X1_2, X2_2, color = 'red')

w = np.random.uniform(low=0, high =3, size=3)

plt.plot([np.min(X1_1), np.max(X1_2)], [np.min(X1_1)*(-w[1])/w[2]+(-w[0]/w[2]),np.max(X1_2)*(-w[1])/w[2]+(-w[0]/w[2])], color = 'black')

X0 = np.ones(2*N)
X1 = np.concatenate((X1_1, X1_2))
X2 = np.concatenate((X2_1, X2_2))
X = np.stack((X0,X1,X2), axis=1)

Y = np.concatenate((Y_1, Y_2))

plt.scatter(X[:,1], X[:,2], c=Y)

epochs = 0
max_epochs = 10000
batch_size = 10

learning_rate = 1e-5

errori = np.zeros(max_epochs)
numero_errori = np.zeros(max_epochs)

while(epochs<max_epochs):
    errori[epochs]=error_w(X,Y,w)
    numero_errori[epochs] = sum(Y*np.dot(X,w)<0)
    rand_perm = np.random.permutation(range(Y.shape[0]))
    X = X[rand_perm,:]
    Y = Y[rand_perm]
    print("Epoca numero: ", epochs)
    print("Errore: ", errori[epochs])
    print("Numero errori: ", numero_errori[epochs]) 
    for j in range(0, Y.shape[0], batch_size):
        X_batch= X[j:j+batch_size,:]
        Y_batch= Y[j:j+batch_size]
        errors = Y_batch - np.dot(X_batch,w)
        w = w+learning_rate*np.dot(X_batch.transpose(), errors)
    epochs+=1
print("L'errore dell'ultima epoca e': ", errori[-2])
print("Il numero di errori dell'ultima epoca e': ", numero_errori[-2])

plt.scatter(X[:,1], X[:,2], c=Y)
plt.plot([np.min(X1_1), np.max(X1_2)], [np.min(X1_1)*(-w[1])/w[2]+(-w[0]/w[2]),np.max(X1_2)*(-w[1])/w[2]+(-w[0]/w[2])], color = 'black')
```

```python
fig,axs = plt.subplots(1,2, figsize=(15,5), sharey=False)
axs[0].plot(range(sum(errori>0)), errori[errori>0])
axs[1].plot(range(sum(errori>0)), numero_errori[errori>0])
```
